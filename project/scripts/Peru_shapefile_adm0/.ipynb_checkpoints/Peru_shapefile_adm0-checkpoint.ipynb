# Using geoBoundaries API

## Peru administrative level 0: `Adm0` (region)

### Loading data

About the geoBoundaries API, the `requests.get` module, requires a three-letter ISO country code `ISO=PER`, and the administrative level number `ADM=ADM0`. 

Then, `dlPath_0` will save a `.geojson` url with the information requested above. 

Finally, `geoBoundary_0` will read the `.geojson` file and return the geometric shape(s) representing the administrative level 0 boundaries from Peru.

See the example below:

```python
r = requests.get("https://www.geoboundaries.org/gbRequest.html?ISO=PER&ADM=ADM0")
dlPath_0 = r.json()[0]['gjDownloadURL']
geoBoundary_0 = requests.get(dlPath_0).json()
# print(geoBoundary["features"][1]["geometry"])
```

Now, as `dlPath_0` is a `.geojson` file, we are saving that file into a `shapefile`. For that task, we import `geopandas` library.

```python
import geopandas as gpd 

gdf_0 = gpd.read_file(dlPath_0)
gdf_0.to_file(r'Peru_adm0.shp')
```

### Visualizing data

```python
r = requests.get("https://www.geoboundaries.org/gbRequest.html?ISO=PER&ADM=ADM0")
dlPath_0 = r.json()[0]['gjDownloadURL']
geoBoundary_0 = requests.get(dlPath_0).json()

#Matplotlib Visualization
fig = plt.figure(1, figsize=(15,15), dpi=100)
axs = fig.add_subplot(111)
axs.set_title('Example Visualization')

xs = []
ys = []
exterior = []
#Accounting for Multipolygon Boundaries
for boundary in geoBoundary_0["features"]:
    if boundary["geometry"]['type'] == "MultiPolygon":
        polys = boundary["geometry"]["coordinates"]
        for poly in polys:
            exterior = poly[0]
            xs, ys = zip(*exterior)
            axs.fill(xs, ys, alpha=0.5, fc='red', ec='black')
    else:
        exterior = boundary["geometry"]["coordinates"][0]
        xs, ys = zip(*exterior)
        axs.fill(xs, ys, alpha=0.5, fc='red', ec='black')

fig.savefig("Peru_adm0.png")
```
